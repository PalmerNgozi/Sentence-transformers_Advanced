```python
# put the tokenizatin method in a function

def tokenizer_function(example):
  return tokenizer(example['sentence1'], example['sentence2'], truncation= True)
```

```python
#in order for us to keep the data as a datset we use the map function 

tokenizer_input = input_data.map(tokenizer_function, batched = True)
tokenizer_input
```

```python
# dynamic padding to put together samples in a batch 

from transformers import DataCollatorWithPadding

collator = DataCollatorWithPadding(tokenizer = tokenizer)
```

```python
sample1 = tokenizer_input["train"][:8]
sample1 = {k: v for k, v in sample1.items() if k not in ["idx", "sentence1", "sentence2"]}
[len(x) for x in sample1["input_ids"]]
```

```python
# batching 
batch1 = collator(sample1)
{k: v.shape for k, v in batch1.items()}
```
